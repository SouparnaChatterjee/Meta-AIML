create a MultiIndex from columns Category and Region

set_index fxn is used here

```python
df=df.set_index(['Category','Region'])
print(df)
```

select all rows from MultiIndex dataframe where Region is North

df.loc() is used for selection

```python
subset=df.loc[df.index.get_level_values('Region')=='North']
print(subset)
```

swap the levels of MultiIndex and sort new first level

.swaplevel and .sortindex are used for sorting first level

```python
df=df.swaplevel().sort_index()
print(df)
```

reset MultiIndex back to regular columns

.reset_index() is used for resetting indices

```python
df=df.reset_index()
print(df)
```

create a pivot table with Region as rows, Category as columns, and sum of values as values

.pivottable is used again

```python
pivot_table=df.pivot_table(index='Region',columns='Category',values='Age',aggfunc='sum')
print(pivot_table)
```

extract the year, month, day from the Date column and cadd them as new columns

```python
df['year']=pd.to_datetime(df['Date']).dt.year
df['month']=pd.to_datetime(df['Date']).dt.month
df['day']=pd.to_datetime(df['Date']).dt.day
print(df)
```

filter rows where date is between "2022-01-01" and "2023-01-01"

```python
filtered_df=df.loc[(df['Date']>'2022-01-01') & (df['Date']<'2023-01-01')]
print(filtered_df)
```

convert a column date with string values like "2022-01-01" into pandas datetime objects

```python
df['date']=pd.to_datetime(df['Date'])
print(df)
```

after grouping by region, filter groups where total sales are >5000

```python
filtered_groups=df.groupby('Region').filter(lambda x:x['Age'].sum()>5000)
print(filtered_groups)
```
